```python
def get_price(symbol,volume, direction = 'BUY'):
    orderbook = fnm.orderbook(symbol)
    dir = 'offer' if direction == 'BUY' else 'bid'
    df = orderbook[dir]
    vol_cumsum = df['volume'].cumsum()
    series = vol_cumsum < volume
    idx = series[series == False].index[0]
    price = df.loc[idx,'price']
    return price 
```
