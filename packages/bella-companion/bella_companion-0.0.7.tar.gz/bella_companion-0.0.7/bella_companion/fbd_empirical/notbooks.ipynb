```python
import json
import os

import polars as pl

log_summary = pl.read_csv(
    os.path.join("../../logs_summaries", "fbd-empirical", "MLP.csv")
)

with open(os.path.join("params", "MLP.json"), "r") as f:
    params = json.load(f)
states = params["types"].split(",")
change_times = (
    pl.read_csv(os.path.join("data", "change_times.csv"), has_header=False)
    .to_series()
    .to_list()
)
```

```python
from phylogenie import load_newick

trees = load_newick("data/trees.nwk")
```

```python
n_time_bins = len(change_times) + 1
for s in states:
    estimates = [
        log_summary[f"birthRatei{i}_{s}_median"].median() for i in range(n_time_bins)
    ]
    plt.step(change_times, estimates[:-1], label=rf"$\lambda_{{{s}}}$")
    plt.legend()
plt.gca().invert_xaxis()  # This reverses the x-axis
```

```python
n_time_bins = len(change_times) + 1
for s in states:
    estimates = [
        log_summary[f"deathRatei{i}_{s}_median"].median() for i in range(n_time_bins)
    ]
    plt.step(change_times, estimates[:-1], label=rf"$\mu_{{{s}}}$")
    plt.legend()
plt.gca().invert_xaxis()  # This reverses the x-axis
```

```python
for s in states:
    estimates = [
        log_summary[f"birthRatei{i}_{s}_median"].median()
        - log_summary[f"deathRatei{i}_{s}_median"].median()
        for i in range(n_time_bins)
    ]
    plt.step(change_times, estimates[:-1], label=rf"$d_{{{s}}}$")
    plt.legend()
plt.gca().invert_xaxis()  # This reverses the x-axis
```
