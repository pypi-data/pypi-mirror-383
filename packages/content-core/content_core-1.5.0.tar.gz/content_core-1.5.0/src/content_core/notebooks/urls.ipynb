```python
from content_core.content.extraction import extract_content

async def process_url(url):
    print("Processing: ", url)
    print("Simple: -------")
    result = await extract_content(dict(url=url, engine="simple"))
    print(result.title[:100])
    print(result.content[:100])
    print("Jina: -------")
    result = await extract_content(dict(url=url, engine="jina"))
    print(result.title[:100])
    print(result.content[:100])
    print("Firecrawl: -------")
    result = await extract_content(dict(url=url, engine="firecrawl"))
    print(result.title[:100])
    print(result.content[:100])
    print("=============================")
```
